```python
from bs4 import BeautifulSoup as soup
from urllib.request import Request,urlopen
import heapq
import string
import re
import nltk
from nltk.corpus import stopwords
from nltk import sent_tokenize,word_tokenize
STOP_WORD=stopwords.words('english')
```

```python
link=input("Enter the URL you want to Scrap:")
```

```python
req=Request(link,headers={'User-Agent':'Mozilla/5.0'})
page=urlopen(req)
bs4page=soup(page,'html.parser')
element=bs4page.find_all('p')
document=''
for i in element:
    document+=(i.text)
type(document)
```

```python

document = document.replace(r'^\s+|\s+?$','')
document = document.replace('\n',' ')
document = document.replace("\\",'')
document = document.replace(",",'')
document = document.replace('"','')
document = re.sub(r'\[[0-9]*\]','',document)
document
```

```python
sentences=sent_tokenize(document)
sentences
```

```python
word_frequency={}
document_for_word=document.lower()
for word in word_tokenize(document_for_word):
    if word not in STOP_WORD and word not in string.punctuation:
        if word not in word_frequency.keys():
            word_frequency[word]=1
        else:
            word_frequency[word]+=1
word_frequency
```

```python
max_frequency=max(word_frequency.values())
for word in word_frequency.keys():
    word_frequency[word]=(word_frequency[word]/max_frequency)
word_frequency
```

```python
sentence_score = {}
for sent in sentences:
    for word in word_tokenize(sent):
        if word in word_frequency.keys():
            if len(sent.split(' '))<100:
                if sent not in sentence_score.keys():
                    sentence_score[sent] = word_frequency[word]
                else:
                    sentence_score[sent]+=word_frequency[word]
sentence_score
```

```python
summary=heapq.nlargest(4,sentence_score,key=sentence_score.get)
```

```python
summary=' '.join(summary)
```
